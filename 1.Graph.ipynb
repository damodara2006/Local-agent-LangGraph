```python
import os
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from langgraph.graph import StateGraph, MessagesState, START, END
from typing import TypedDict
from IPython.display import Image, display
load_dotenv()


llm = ChatGroq(  model="llama-3.1-8b-instant",
    temperature=0.0,
    max_retries=2
    )

class graph_schema(TypedDict):
    name : str
    message : str

# print(llm.invoke("Hii"))
def greet(state : graph_schema) -> graph_schema:
    curr_name = state["name"]
    curr_message = state["message"]
    
    res = llm.invoke(f"My name is {curr_name}. {curr_message}").content
    
    state["message"] = f"Your messages is {curr_message}. My res is {res}"
    
    return state

first_graph = StateGraph(graph_schema)

first_graph.add_node("greet", greet)

first_graph.add_edge(START, "greet")
first_graph.add_edge("greet", END)

graph = first_graph.compile()

Image(graph.get_graph().draw_mermaid_png())
```
